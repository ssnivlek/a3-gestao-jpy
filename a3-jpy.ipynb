```python
import matplotlib.pyplot as plt
import textwrap
```

```python
def calcular_porcentagem_acumulada(valores):
    total = sum(valores)
    porcentagem_acumulada = 0
    porcentagens = []
    
    for valor in valores:
        porcentagem = (valor / total) * 100
        porcentagem_acumulada += porcentagem
        porcentagens.append(porcentagem_acumulada)
    
    return porcentagens
```

```python
categorias = ['Sabor ruim', 'Falta de gás', 'Objeto Estranho', 'Propaganda Enganosa', 'Promoções', 'Embalagens Retornáveis']
reclamacoes = [3334, 1786, 1248, 1201, 979, 567]
percentages_acumuladas = calcular_porcentagem_acumulada(reclamacoes)


fig, ax1 = plt.subplots()

ax1.bar(categorias, reclamacoes, color='slateblue')
ax1.set_xlabel('Categorias')
ax1.set_ylabel('Número de reclamações', color='black')
ax1.tick_params('y', colors='black')


ax2 = ax1.twinx()
ax2.plot(categorias, percentages_acumuladas, 'ro-', color='darkred')
ax2.set_ylabel('Porcentagem acumulada de reclamações', color='darkslategrey')
ax2.tick_params('y', colors='darkslategrey')



# Ajustar rótulos do eixo x com quebra de linha
plt.title('Gráfico de reclamações por categoria')
ax1.set_xticklabels([textwrap.fill(label, 11) for label in categorias], rotation=45)


plt.subplots_adjust(bottom=0.2)

fig.tight_layout()
plt.show()
```
